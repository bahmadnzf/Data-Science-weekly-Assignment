```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ds_salaries=pd.read_csv('ds_salaries.csv')
ds_salaries.head()
```

```python
ds_salaries.info()
```

```python
plt.hist(ds_salaries['salary_in_usd'], bins=10, edgecolor='brown', color='black')
plt.title('Salary Distribution(USD)')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(ds_salaries['experience_level'], edgecolor='black', color='gold')
plt.title('Experience Distributions')
plt.xlabel('Experience')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(ds_salaries['company_size'], edgecolor='black', color='red')
plt.title('Company Size Distribution')
plt.xlabel('company size')
plt.ylabel('Frequency')
plt.show()
```

```python
print(f'Experience levels: {ds_salaries['experience_level'].unique()}')
print(f'Company Size: {ds_salaries['company_size'].unique()}')
```

```python
import seaborn as sns
sns.set(style='whitegrid')
sns.scatterplot(data=ds_salaries, x='salary_in_usd', y='experience_level', hue='company_size', alpha=0.8, palette='colorblind')

plt.title('Salary Distribution(USD)')
plt.xlabel('Salary in USD')
plt.ylabel('Experience Level')
plt.legend(title='Company Size')
plt.show()
```

The visualization shows that higher experience levels generally correspond to higher salaries, regardless of company size.
However, larger companies('L') tend to offer higher salaries across all experience levels compared to medium('M') and small('S')companies. This suggests that both experience and company size play significant roles in determining salary levels in data science roles.
