```python
import matplotlib.pyplot as plt
import pandas as pd 

```

```python
birds = pd.read_csv('birds.csv')
birds.head()
```

```python
birds.info()
```

```python
snow_goose = birds[birds['Name'] == 'Snow goose'].iloc[0]
snow_goose 
```

```python
avg_length=(snow_goose['MinLength']+ snow_goose['MaxLength'])/2
avg_mass=(snow_goose['MinBodyMass']+ snow_goose['MaxBodyMass'])/2
avg_wingspan=(snow_goose['MinWingspan']+ snow_goose['MaxWingspan'])/2

plt.bar(['Length', 'Body Mass', 'Wingspan'], [avg_length,avg_mass, avg_wingspan], color=['Brown','Gold', 'Darkblue'])
plt.title('Average Size Attributes of Rock Pigeon')
plt.ylabel('Measurement(CM)')
plt.show()

```

```python
min_values=[snow_goose['MinLength'],snow_goose['MinBodyMass'],snow_goose['MinWingspan']]
max_values=[snow_goose['MaxLength'],snow_goose['MaxBodyMass'],snow_goose['MaxWingspan']]
labels=['Length','Body Mass','Wingspan']

plt.plot(labels,min_values, marker='v', label='Min')
plt.plot(labels,max_values, marker='v', label='Max')
plt.title('Min vs Max Size Measurement of Snow Goose')
plt.ylabel('Measurement(CM)')
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```

```python
x=[snow_goose['MinBodyMass'], snow_goose['MaxBodyMass']]
y=[snow_goose['MinWingspan'],snow_goose['MaxWingspan']]
labels=['Min', 'Max']

plt.scatter(labels, x,y, color='orange')

plt.title('Wingspan vs Body Mass--Snow Goose')
plt.xlabel('Body Mass')
plt.ylabel('Wingspan')
plt.grid(True, alpha=0.5)
plt.show()
```
