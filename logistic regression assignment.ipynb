```python
#Importing the Necessary Libraries
import pandas as pd
import numpy as np

df =pd.read_csv("Telco_Customer_Churn_Dataset.csv", delimiter = ',')#Reading my Dataset to my editor
df#Displaying the Dataset
```

```python
#Grouping the dataset into Categorical and NUmerical 
cat = []#for Categorrical Data
num = [] #for Numerical Data
for x in df.columns:
     if df[x].dtypes == object:
      cat.append(x)
     else:
         num.append(x)
cat
```

```python
cat_columns=['customerID','gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection', 'TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','TotalCharges','Churn']
```

```python
df.drop('customerID', axis=1, inplace=True)#Droping The CustomerID because it is unneccesary for the Model 
```

```python
#Encoding my Binary Data(Those with 2 Values)
binary_cols=['gender', 'Partner','Dependents','PhoneService','PaperlessBilling','Churn']
for col in binary_cols:
    df[col]=df[col].map({'Yes':1, 'No':0, 'Male': 1,'Female': 0})
```

```python
#Displaying the information of my Dataset
df.info()
```

```python
#Encoding the Multi-Value Columns 
multi_class_cols=['MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod']
df=pd.get_dummies(df, columns=multi_class_cols, drop_first=True)
```

```python
import re
```

```python
#Changing the Data type of TotalCharges 
total = []
for i in df['TotalCharges'] :
    if i.isdigit() == False :
        total.append(0)
    else :
        total.append(i)
```

```python
#Overiding the Values of the TotalCharges 
df['TotalCharges'] = total
```

```python
#Changed Data type of the TotalCharges
df['TotalCharges'] = df['TotalCharges'].astype('float64')
```

```python
#Displaying the Data info to see the Changes are effected
df.info()
```

```python
#Getting my Data for Model Selection
target_column="Churn"
X=df.drop(columns=[target_column])
y=df[target_column]
```

```python
#Importing the neccesary library for model Selection 
from sklearn.model_selection import train_test_split
```

```python
#Dividing my Data for Model Training and Testing
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.20, random_state= 32, stratify=y)
```

```python
#IMporting the necessary libraries for Model prediction Algorithms 
from sklearn.linear_model import  LogisticRegression
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
m =LogisticRegression(random_state=42)
m.fit(X_train, y_train)
```

```python
#Displaying the Distribution of Training and Testing data
print('Class distribution in y_train:',np.bincount(y_train))
print('Class distribution in y_test:',np.bincount(y_test))
```

```python
#Model Prediction Using LongisticRegression Algorithm
m_pred= m.predict(X_test)
m_pred
```

```python
#Checking the Certainty of the Prediction
m.predict_proba(X_test)
```

```python
#Checking the metrics of my Model 
from sklearn.metrics import accuracy_score, classification_report, precision_score, f1_score, recall_score
```

```python
#Matrix Variables Definition
def matrics(y_test,m_pred):
    acc = accuracy_score(y_test, m_pred)
    f1 = f1_score(y_test, m_pred, pos_label=True)
    pr = precision_score(y_test, m_pred, pos_label=True)
    rc = recall_score(y_test, m_pred, pos_label=True)
    return (acc, f1, pr, rc)    
```

```python
#Matrix variables assignment
m_acc, m_f1, m_pr, m_rc = matrics(y_test, m_pred)
```

```python
#Matrix showing each actual vs predicted class
pd.crosstab(y_test, m_pred)
```

```python
score= [m_acc,m_acc, m_f1, m_f1, m_pr, m_pr, m_rc, m_rc]
def met_df(scores):
    alg = ['M','N','M','N','M','N', 'M','N']
    met = ['Accuracy', 'Accuracy', 'F1_score', 'F1_score', 'Precision_score', 'Precision_score', 'Recall_score','Recall_score']
    base_met = pd.DataFrame({"Matrices": met, "Algorithm": alg, "Score": scores})
    return base_met
```

```python
#Showing the Evaluation metrics of the LongicalRegression algorithm
base_matrics= met_df(score)
base_matrics
```

```python
#Balancing the Data for better Model prediction
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
X_train_resampled, y_train_resampled=smote.fit_resample(X_train,y_train)
```

```python
#Using the XGboost for better imbalance Data Prediction 
from xgboost import XGBClassifier
import numpy as np
xgb_model=XGBClassifier(scale_pos_weight=1.8, random_state=42)

xgb_model.fit(X_train_resampled, y_train_resampled)
y_probs_xgb=xgb_model.predict_proba(X_test)[:, 1]
new_threshold=0.5
y_pred_adjusted_xgb=(y_probs_xgb>=new_threshold).astype(int)
print(classification_report(y_test,y_pred_adjusted_xgb))
```

```python
#Matrix Variables Definitions
def matrics(y_test,y_pred_adjusted_xgb):
    acc = accuracy_score(y_test, y_pred_adjusted_xgb)
    f1 = f1_score(y_test, y_pred_adjusted_xgb, pos_label=True)
    pr = precision_score(y_test, y_pred_adjusted_xgb, pos_label=True)
    rc = recall_score(y_test, y_pred_adjusted_xgb, pos_label=True)
    return (acc, f1, pr, rc)  

m_acc, m_f1, m_pr, m_rc = matrics(y_test, y_pred_adjusted_xgb)
```

```python
#Showing the Evaluation metrics of the XGboost algorithm
score= [m_acc,m_acc, m_f1, m_f1, m_pr, m_pr, m_rc, m_rc]
def alg_df(scores):
    alg = ['M','N','M','N','M','N', 'M','N']
    met = ['Accuracy', 'Accuracy', 'F1_score', 'F1_score', 'Precision_score', 'Precision_score', 'Recall_score','Recall_score']
    base_met = pd.DataFrame({"Matrices": met, "Algorithm": alg, "Score": scores})
    return base_met
base_matrics= alg_df(score)
base_matrics
```
